```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load the dataset
file_path = "hapiscore_whr.csv"  # Ensure this file is in the same directory as the notebook
df = pd.read_csv(file_path)
```

```python
# Transpose the dataset for easier plotting
df_melted = df.melt(id_vars=["country"], var_name="Year", value_name="Happiness Score")
df_melted.dropna(inplace=True)  # Remove missing values
df_melted["Year"] = df_melted["Year"].astype(int)  # Convert Year to integer
```

```python
# Select a subset of countries to keep the chart readable
selected_countries = ["United States", "Canada", "Germany", "India", "China", "Brazil", "Australia"]
df_filtered = df_melted[df_melted["country"].isin(selected_countries)]
```

```python
# Plotting
plt.figure(figsize=(10, 6))
for country in selected_countries:
    country_data = df_filtered[df_filtered["country"] == country]
    plt.plot(country_data["Happiness Score"], country_data["Year"], marker='o', linestyle='-', label=country)
```

```python
# Labels and title
plt.xlabel("Happiness Score (WHR)")
plt.ylabel("Year")
plt.title("Happiness Score Changes Over the Years")
plt.legend(title="Country")
plt.grid(True)

# Show the plot
plt.show()
```
